```python
#Exercício Python 10: Crie um programa que leia quanto dinheiro uma pessoa tem na carteira e mostre quantos dólares ela pode comprar.
import requests
class ConversorEmTempoReal():
    def __init__(self,url):
        self.data= requests.get(url).json()
        self.currencies = self.data['rates']
    def convert(self, from_currency, to_currency, amount): 
        initial_amount = amount 
        #Primeiro é feita a conversão caso não esteja em dollar
        #Uma vez que nossa base é em dollar
        if from_currency != 'USD' : 
            amount = amount / self.currencies[from_currency] 

        # Limitando a precisão para 4 casas decimais
        amount = round(amount * self.currencies[to_currency], 4) 
        return amount

url = 'https://api.exchangerate-api.com/v4/latest/USD'
converter = ConversorEmTempoReal(url)
saldo= float(input('Entre com o seu saldo atual: '))
saldousd= converter.convert('BRL','USD',saldo)
print(f'Com R$ {saldo} você pode comprar $ {saldousd}')
```
